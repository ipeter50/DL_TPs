## 1. Loading and normalizing CIFAR10

Using ``torchvision``, it’s extremely easy to load CIFAR10.

````python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import mpld3
mpld3.enable_notebook()

import time

import torch
import torchvision
import torchvision.transforms as transforms
````

The output of torchvision datasets are PILImage images of range [0, 1].
We transform them to Tensors of normalized range [-1, 1]

**NB** Modify the code below to only use a small part of the dataset if your computer is very slow.

````python
transform = transforms.Compose(
    [transforms.RandomHorizontalFlip(), # data augmentation: flip horizontally some images in dataset, reduce overfitting
    # transforms.RandomRotation(15), # tried to add also random rotation, but results in lower accuracy
     transforms.ToTensor(),
     transforms.Normalize((0.5, 0.5, 0.5),
                          (0.5, 0.5, 0.5))]
)


# Load dataset
trainset = torchvision.datasets.CIFAR10(root='./data', train=True,
                                        download=True, transform=transform)

testset = torchvision.datasets.CIFAR10(root='./data', train=False,
                                       download=True, transform=transform)


classes = ('plane', 'car', 'bird', 'cat', 'deer',
           'dog', 'frog', 'horse', 'ship', 'truck')

used_categories = range(len(classes))


#Define a sampler to get a validation 
indices = [i for i in range(50000)]
split = 1000

train_idx, validation_idx = indices[split:], indices[:split]

train_sampler = torch.utils.data.SubsetRandomSampler(train_idx)
validation_sampler = torch.utils.data.SubsetRandomSampler(validation_idx)

trainloader = torch.utils.data.DataLoader(trainset, batch_size=128,
                                          sampler=train_sampler, shuffle=False, num_workers=2)
validationloader = torch.utils.data.DataLoader(trainset, batch_size=128,
                                          sampler=validation_sampler, shuffle=False, num_workers=2)

testloader = torch.utils.data.DataLoader(testset, batch_size=128,
                                         shuffle=True, num_workers=2)
train_data_iter = iter(trainloader)
test_data_iter = iter(testloader)
print('used classes:', classes)
````

````python
print("Training data")
print(trainset.train_data.shape)
print(len(trainset.train_labels))
print()


print("Test data")
print(testset.test_data.shape)
print(len(testset.test_labels))
print()
````

Let us show some of the training images, for fun.



````python
# Run this cell multiple time to see more samples

def imshow(img):
    """ show an image """
    img = img / 2 + 0.5 # unnormalize
    npimg = img.numpy()
    plt.imshow(np.transpose(npimg, (1, 2, 0)))

# get some random training images
images, labels = train_data_iter.next()

# show images
imshow(torchvision.utils.make_grid(images[:4]))

# print labels
print(' '.join('%5s' % classes[labels[j]] for j in range(4)))
````

## 2. Define a Convolutional Neural Network

**Assignment 1:** Define a convolutional neural network. 
You may use the code from previous notebooks.
We suggest that you start with a small network, and make sure that everything is working.
Once you can train successfully come back and improve the architecture

````python
from torch.autograd import Variable
import torch.nn as nn
import torch.nn.functional as F


channels = 3
nb_filter_conv1 = 96
kernel_size_conv1 = 5
padding_conv1 = 2
stride_conv1 = 1

kernel_size_pool1 = 2

nb_filter_conv2 = 256

nb_filter_conv3 = 384
kernel_size_conv3 = 3
padding_conv3 = 1

l_1_features = 1024
l_2_features = 512
l_3_features = 256

l_out_features = 10

class Net(nn.Module):
    def __init__(self, num_classes):
        super(Net, self).__init__()
        self.num_classes = num_classes
        
        # first convolution layer + max pooling
        self.conv_1 = nn.Conv2d(in_channels= channels,
                                out_channels= nb_filter_conv1,
                                kernel_size= kernel_size_conv1,
                                stride= stride_conv1,
                                padding= padding_conv1)
        self.max_pool_1 = nn.MaxPool2d(kernel_size= kernel_size_pool1)
        
        # add batchnorm to the convolution
        self.batch_norm_conv1 = nn.BatchNorm2d(nb_filter_conv1)
        
        # second convolution layer 
        self.conv_2 = nn.Conv2d(in_channels= nb_filter_conv1,
                                out_channels= nb_filter_conv2,
                                kernel_size= kernel_size_conv1,
                                stride= stride_conv1,
                                padding= padding_conv1)
        self.max_pool_2 = nn.MaxPool2d(kernel_size= kernel_size_pool1)
        
        # adding batchnorm
        self.batch_norm_conv2 = nn.BatchNorm2d(nb_filter_conv2)
        
        # third convolution
        self.conv_3 = nn.Conv2d(in_channels= nb_filter_conv2,
                                out_channels= nb_filter_conv3,
                                kernel_size= kernel_size_conv3,
                                stride= stride_conv1,
                                padding= padding_conv3)
        
        # adding batchnorm
        self.batch_norm_conv3 = nn.BatchNorm2d(nb_filter_conv3)
        
        self.max_pool_3 = nn.MaxPool2d(kernel_size= kernel_size_pool1, stride = 1) #only stride of 1, keep feature
        
        
        self.out_conv_features = 384 * 7 * 7 # number of feature remaining
        
        # first linear layer
        self.linear_1 = nn.Linear(in_features= self.out_conv_features,
                                  out_features= l_1_features,
                                  bias = True)
        
        # adding batchnorm
        self.batch_norm_l1 = nn.BatchNorm1d(l_1_features)
        
        # second linear layer
        self.linear_2 = nn.Linear(in_features= l_1_features,
                                  out_features= l_2_features,
                                  bias = True)
        
        # adding batchnorm
        self.batch_norm_l2 = nn.BatchNorm1d(l_2_features)
        
        # third linear layer
        self.linear_3 = nn.Linear(in_features= l_2_features,
                                  out_features= l_3_features,
                                  bias = True)
        
        # adding batchnorm
        self.batch_norm_l3 = nn.BatchNorm1d(l_3_features)
        
        # output layer
        self.linear_out = nn.Linear(in_features= l_3_features,
                                  out_features= l_out_features,
                                  bias = False)

    def forward(self, x):
        
        
        x = F.relu(self.batch_norm_conv1(self.conv_1(x)))
        x = self.max_pool_1(x)
        
        x = F.relu(self.batch_norm_conv2(self.conv_2(x)))
        x = self.max_pool_2(x)
        
        x = F.relu(self.batch_norm_conv3(self.conv_3(x)))
        x = self.max_pool_3(x)

        x = x.view(-1, self.out_conv_features) # flatten feature to match with next layer
        
        x = F.relu(self.batch_norm_l1(self.linear_1(x)))
        
        x =F.relu(self.batch_norm_l2(self.linear_2(x)))

        x = F.relu(self.batch_norm_l3(self.linear_3(x)))
        
        # softmax on the output layer
        x = F.softmax(self.linear_out(x)) 
        
        return x
    

net = Net(len(used_categories)).cuda()
print(net)
````

## 3. Define a Loss function and optimizer

**Assignment 2:** Implement the criterion and optimizer. 
We suggest Classification Cross-Entropy loss and SGD with momentum.
You might need to experiment a bit with the learning rate.

````python
import torch.optim as optim

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(params = net.parameters(), lr= 0.01, momentum = 0.9, weight_decay = 1e-5)
````

## 4. Train the network

**Assignment 3:** Finish the training loop below. 
Start by using a small number of epochs (e.g. 3).
Even with a low number of epochs you should be able to see results that are better than chance.
When everything is working increase the number of epochs to find out how good your network really is.

````python
num_epoch = 100

accuracy= []
for epoch in range(num_epoch):  # loop over the dataset multiple times
    time_epoch = time.time()
    running_loss = 0.0
    net.train() # nable batchnorm and dropout
    for i, data in enumerate(trainloader, 0):
        # get the inputs
        inputs, labels = data

        # wrap them in Variable
        inputs, labels = Variable(inputs).cuda(), Variable(labels).cuda()
        outputs = net(inputs)
        
        # zero the parameter gradients
        optimizer.zero_grad()
        

        # forward + backward + optimize
        batch_loss = criterion(outputs, labels)
        batch_loss.backward()
        optimizer.step()
        
        # print statistics
        running_loss += batch_loss
        
    net.eval()
    
    #Compute network accuracy on validation set
    correct = 0
    for i, data in enumerate(validationloader, 0):
        #get data from validation set and wrap them in Variable + send them to CUDA
        validation_inputs, validation_labels = data
        validation_inputs, validation_labels = Variable(validation_inputs).cuda(), Variable(validation_labels).cuda()

        #compute the outputs
        validation_outputs = net(validation_inputs)
        
        #count the right answers
        _, predicted = torch.max(validation_outputs.data, 1)
        correct += (predicted == validation_labels).sum()
        
    # print metrics every epoch
    acc= int(correct)/1000.0
    
    print('Epoch %d: Loss: %.3f Accuracy: %.3f' %
          (epoch + 1, running_loss / 1000, acc))
    
    print("Time spent: %f" % (time.time()-time_epoch))
    
    accuracy.append(acc)
    correct = 0
    running_loss = 0.0

plt.figure()
plt.plot([i for i in range(num_epoch)], accuracy)
plt.legend(['Train Accucary'])
plt.xlabel('Epochs'), plt.ylabel('Acc')
print('Finished Training')
````

## 5. Test the network on the test data

Now we need to check if the network has learnt anything at all.
We will check this by predicting the class label that the neural network outputs, and checking it against the ground truth.
If the prediction is correct, we add the sample to the list of correct predictions.

Okay, first step. Let us display an image from the test set to get familiar.

````python
images, labels = test_data_iter.next()

# print images
imshow(torchvision.utils.make_grid(images[:4]))
plt.show()

outputs = net(images.cuda())

print('GroundTruth:  ', ' '.join('%5s' % classes[labels[j]] for j in range(4)))

_, predicted = torch.max(outputs.data.cpu(), 1)
print('Predicted:    ', ' '.join('%5s' % classes[predicted[j]] for j in range(4)))
````

Let us look at how the network performs on the whole dataset.

````python
correct = 0
total = 0

for data in testloader:
    images, labels = data
    outputs = net(Variable(images.cuda()))
    _, predicted = torch.max(outputs.data, 1)
    total += labels.size(0)
    correct += (predicted.cpu() == labels).sum()
print('Accuracy of the network on the {} test images: {:4.2f} %'.format(
    testset.test_data.shape[0], 100.0 * float(correct) / total))
````

Hopefully the network is better than chance, which is $\frac{1}{\text{number of classes}}$ accuracy (randomly picking
a class).


We can also examine which class the network found the most difficult (makes more sense if you have many clases):

````python
class_total = list(0. for i in range(len(classes)))
class_correct = list(0. for i in range(len(classes)))

for data in testloader:
    images, labels = data
    outputs = net(Variable(images.cuda()))
    _, predicted = torch.max(outputs.data, 1)
    c = (predicted.cpu() == labels).squeeze()
    
    for i in range(len(c)):
        label = labels[i]
        class_correct[label] += c[i].numpy()
        class_total[label] += 1

for i in range(len(classes)):
    print('Accuracy of {:5s} : {:5.2f} %'.format(
        classes[i], 100 * class_correct[i] / class_total[i]))
````

* **Monotonicity of softmax**  
  We to show that $\frac{\partial a_j^L}{\partial z_k^L}$ is positive if $j = k$ and negative otherwise.
  
  - If $j=k$:  
$$\begin{gather*}     
            \frac{\partial a_j^L}{\partial z_k^L} = \frac{\partial}{\partial z_k^L} \frac{e^{z_k^L}}{\sum\limits_i e^{z_i^L}}  
            = \frac{e^{z_k^L}\sum\limits_i e^{z_i^L} - e^{z_k^L} \frac{\partial}{\partial z_k^L}\sum\limits_i e^{z_i^L}}{(\sum\limits_i e^{z_i^L})^2}
            = \frac{e^{z_k^L} \sum\limits_{i \neq k} e^{z_i^L}}{(\sum\limits_i e^{z_i^L})^2} > 0
  \end{gather*}$$  
  - If $j \neq k$:  
   
   $$\begin{gather*}     
            \frac{\partial a_j^L}{\partial z_k^L} =  e^{z_j^L}\frac{\partial}{\partial z_k^L} \frac{1}{\sum\limits_i e^{z_i^L}}
            = - e^{z_j^L}\frac{e^{z_k^L}}{(\sum\limits_i e^{z_i^L})^2} < 0
  \end{gather*}$$
  
  Hence, we have shown that increasing $z_j^L$ will increase $a_j^L$ and decrease all the other output activations. 
  
* **Non-locality of softmax**  

With sigmoid layers, the output $a_j^L$ is a function of the corresponding weighted input: $a_j^L = \sigma(z_j^L)$.
With softmax, it is not the case anymore: $a_j^L= \frac{e^{z_j^L}}{\sum\limits_i e^{z_i^L}}$ is a function of all the weighted input of the layer. This can be explained because softmax layers build a probability distribution based on the values of its inputs. Then, to be a probability distribution, the sum of the outputs must be equal to $1$ and for that each output can only be calculated given all the inputs.
